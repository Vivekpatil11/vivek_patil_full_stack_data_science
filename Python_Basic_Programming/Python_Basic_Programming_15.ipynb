```python
#1
def number(n):
    for i in range(n):
        if(i%5==0 and i%7==0):
            yield i
lst=[]
n=int(input("Enter the value :"))
for i in number(n):
    lst.append(str(i))
print(','.join(lst))
```

```python
#2
def evn_num(n):
    for i in range(n+1):
        if(i%2==0):
            yield i
n=int(input("Enter the number"))
lst=[]
for i in evn_num(n):
    lst.append(str(i))
print(",".join(lst))
```

```python
#3
def fib(n):
    i=0
    j=1
    for k in range(n+1):
        yield i
        i,j=j,i+j
n=int(input("Enter the number"))
print(",".join([str(x) for x in fib(n)]))
```

```python
#5
class shape:
    def area(self):
        return 0
class square(shape):
    def __init__(self,length):
        self.length=length
    def area(self):
        return self.length**2

shp=shape()
print(shp.area())
sqr=square(4)
print(sqr.area())
```
