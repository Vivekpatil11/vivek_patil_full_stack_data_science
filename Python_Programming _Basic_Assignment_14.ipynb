```python
#1
def multiple_7(n):
    for i in range(1,n):
        if(i%7==0):
            yield i
n=int(input("Enter the value :"))
for i in multiple_7(n):
    print(i)
```

```python
#2
sentence = input("Enter a sentence: ").split(' ')
dct = {}
for c in sentence:
    if c in dct.keys():
        dct[c] += 1
    else:
        dct[c] = 1

for item in sorted(dct.items()):
    print(item[0],':',item[1])
```

```python
#3 
class person:
    def getgender(self):
        pass
class male(person):
    def getgender(self):
        print('male')
class female(person):
    def getgender(self):
        print('female')
Male=male()
Female=female()
Male.getgender()
Female.getgender()
```

```python
#4
subject=["I", "You"]
verb=["Play", "Love"]
obj=["Hockey","Football"]
sentence = []
for i in range(2):
    for j in range(2):
        for k in range(2):
            sentence.append(subject[i] + " " + verb[j] + " " + obj[k])
for i in sentence:
    print(i)
```

```python
def binary_search(lst,n):
    low=0
    high=len(lst)-1
    mid=0
    while(high>=low):
        mid=(high+low)//2
        if(lst[mid]>n):
            high=mid-1
        elif(lst[mid]<n):
            low=mid+1
        else:
            return mid
    return -1
lst=[12,24,36,48,60,72]
print(binary_search(lst,36))
```
