```python
#4
num=371
val=num
res=0
while(num):
    rem=num%10
    num=num//10
    res=res+rem**len(str(val))
if(res==val):
    print("The given value is the armstrong number")
else:
    print("The given value is not the armstrong number")
```

```python
#5
lower=100
upper=2000
for num in range(lower,upper):
    order=len(str(num))
    val=num
    res=0
    while(num):
        rem=num%10
        num=num//10
        res=res+rem**order
    if(res==val):
        print(res)
```
