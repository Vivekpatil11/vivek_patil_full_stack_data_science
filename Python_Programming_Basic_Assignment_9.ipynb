```python
#2
def check_disarium(num):
    length=len(str(num))
    exp=num
    res=0
    while(num>0):
        rem=num%10
        res+=(rem**length)
        num=num//10
        length-=1
    if(res==exp):
        return True
    else:
        return False
for i in range(100):
    if(check_disarium(i)):
        print(i)
```
