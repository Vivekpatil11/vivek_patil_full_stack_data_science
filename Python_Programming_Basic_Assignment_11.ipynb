```python
#5
a="Vivek patil Bijapur Kolar Venu PES"
b="Vivek PES Gownipalli Srinivaspur"
a=a.split()
b=b.split()
did={}
for i in a:
    if i not in did:
        did[i]=0
    did[i]+=1
for i in b:
    if i not in did:
        did[i]=0
    did[i]+=1
print([word for word in did if(did[word]==1)])
```
