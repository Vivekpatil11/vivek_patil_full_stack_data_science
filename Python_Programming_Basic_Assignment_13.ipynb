```python
#1
import math
c=50
h=30
d=[int(x) for x in input().split(',')]
for i in d:
    q=math.sqrt((2*c*i)/h)
    print(int(q))
```

```python
#2
x=int(input("Enter the number of rows :"))
y=int(input("Enter the number of columns :"))
res=[]
for i in range(x):
    res.append([])
    for j in range(y):
        res[i].append(i*j)
print(res)
```

```python
#5
sentence=input()
letters=0
numbers=0
for i in sentence:
    if(i.isdigit()):
        numbers+=1
    if(i.isalpha()):
        letters+=1
print("no of letters :",letters)
print("no of numbers :",numbers)
```

```python
#6
passwords=list(input().split(','))
res=[]
for i in passwords:
    if(len(i)>=6 and len(i)<=12):
        check=[0,0,0,0]
        for j in i:
            if('a'<=j and 'z'>=j):
                check[0]=1
            elif('A'<=j and 'B'>=j):
                check[1]=1
            elif('0'<=j and '9'>=j):
                check[2]=1
            else:
                check[3]=1
        if(sum(check)==4):
            res.append(i)
print(','.join(res))
```
