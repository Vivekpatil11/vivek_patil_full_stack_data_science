```python
#1
def fib(n):
    if(n<=1):
        return n
    else:
        return fib(n-1)+fib(n-2)
for i in range(10):
    print(fib(i))
```

```python
#2
def facto(n):
    if(n<=1):
        return 1
    else:
        return n*facto(n-1)
print(facto(5))
```
